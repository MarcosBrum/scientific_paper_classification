```python
import json

import matplotlib.pyplot as plt
import seaborn as sns
```

## 0.2 Data

```python
DATA_PATH = "../archive/arxiv-metadata-oai-snapshot.json"
```

```python
with open(DATA_PATH, "r") as f:
    metadata = f.readlines()

metadata = [json.loads(data) for data in metadata]
metadata[0]
```

# 1 Count categories

```python
categories = [doc['categories'] for doc in metadata]
categories_flat = [_c for c in categories for _c in c.split(' ')]
```

```python
plt.figure(figsize=(20, 15))

sns.countplot(x=categories_flat, stat='count')
plt.xticks(rotation=270)

plt.show()
plt.close()
```

# 2 Papers in Computer Science

```python
cs = [doc for doc in metadata if "cs." in doc["categories"]]

cs_cat = [doc['categories'] for doc in cs]
cs_cat_flat = [_c for c in cs_cat for _c in c.split(' ')]
cs_cat_flat = [c for c in cs_cat_flat if c.startswith("cs.")]
```

```python
plt.figure(figsize=(20, 15))

sns.countplot(x=cs_cat_flat, stat='count')
plt.xticks(rotation=270)

plt.show()
plt.close()
```
