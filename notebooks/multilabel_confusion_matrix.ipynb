```python
import sys
from os.path import join

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sys.path.append("../")
from source.utils import CATEGORIES, MODELS_PATH
```

# 1 Confusion Matrices

```python
with open(join(MODELS_PATH, "multilabel_confusion_matrix.npy"), "rb") as f:
    confusion_matrices = np.load(f)
```

```python
for category, conf_mt in zip(CATEGORIES, confusion_matrices):
    plt.figure(figsize=(8, 6))
    plt.title(category)
    sns.heatmap(data=conf_mt, annot=True,
                cmap=plt.cm.Blues, fmt='', cbar=False,
                xticklabels=["Pred True", "Pred False"],
                yticklabels=["Annot True", "Annot False"])
    plt.show()
    plt.close()
```
